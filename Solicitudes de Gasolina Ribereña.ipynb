```python
sol_proc = solicitudes[['ano', 'solicitudes_recibidas', 'solicitudes_atendidas']].groupby(['ano']).sum()

sol_proc['porcentaje'] = sol_proc['solicitudes_atendidas']*100/sol_proc['solicitudes_recibidas']

sol_proc.head()

# Solicitudes recibidas, procesadas y porcentaje de solicitudes atendidas por año por estado
```
