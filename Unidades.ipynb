```python
%matplotlib inline

import pandas

unidades = pandas.read_csv('datos/unidades_economicas.csv')

unidades.groupby(['tipo']).size().plot.pie()
```

```python
unidades.groupby(['estado']).size().plot.bar()
```

```python
unidades.groupby(['tipo_persona']).size().plot.pie()
```

```python
iunidades = unidades.set_index(['rnpa'])

beneficiarios_menores = pandas.read_csv('datos/beneficiarios_menores.csv') 
beneficiarios_componentes = pandas.read_csv('datos/beneficiarios_componentes.csv')                   
beneficiarios_gasolina = pandas.read_csv('datos/beneficiarios_gasolina.csv')                      
beneficiarios_diesel = pandas.read_csv('datos/beneficiarios_diesel.csv')                        
beneficiarios_infraestructura = pandas.read_csv('datos/beneficiarios_infraestructura.csv')              
beneficiarios_electricos = pandas.read_csv('datos/beneficiarios_electricos.csv')                                         
beneficiarios_motores = pandas.read_csv('datos/beneficiarios_motores.csv')
```

```python
# Cuantos RNPA de beneficiarios de motores tenemos en nuestras unidades

{
'menores': len(pandas.merge(beneficiarios_menores, iunidades, left_on=['rnpa'], right_index=True)),
'componentes': len(pandas.merge(beneficiarios_componentes, iunidades, left_on=['rnpa'], right_index=True)),
'gasolina': len(pandas.merge(beneficiarios_gasolina, iunidades, left_on=['rnpa'], right_index=True)),
'diesel': len(pandas.merge(beneficiarios_diesel, iunidades, left_on=['rnpa'], right_index=True)),
'infraestructura': len(pandas.merge(beneficiarios_infraestructura, iunidades, left_on=['rnpa'], right_index=True)),
'electricos': len(pandas.merge(beneficiarios_electricos, iunidades, left_on=['rnpa'], right_index=True)),
'motores': len(pandas.merge(beneficiarios_motores, iunidades, left_on=['rnpa'], right_index=True))
}
```
