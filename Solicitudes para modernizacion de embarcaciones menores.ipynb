```python
%matplotlib inline

import pandas

solicitudes = pandas.read_csv('datos/solicitudes_embarcaciones_menores.csv')
```

```python
sol_proc = solicitudes[['A単o', 'Registradas', 'Apoyadas']].groupby(['A単o']).sum()

sol_proc['porcentaje'] = sol_proc['Apoyadas']*100/sol_proc['Registradas']

sol_proc.plot.pie(y='A単o')

# Solicitudes recibidas, procesadas y porcentaje de solicitudes atendidas por a単o por estado
```
